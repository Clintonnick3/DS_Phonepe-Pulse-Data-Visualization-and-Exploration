```python
#Importing necessary libraries

import pandas as pd                  # For creating dataframes.
import mysql.connector as sql        # To Connect with Mysql Database.
import streamlit as st               # To create Interactive web applications.
import plotly.express as px          # For creating interactive and expressive visualizations.
import os                            # For interaction with operating system.
import json                          # To work with Json format
from PIL import Image                # Python Imaging Library (PIL) and is used for working with images in Python.
```

```python
# Defining the path to the directory containing the files.

path = "D:/Clinton files.py/My details/Fin_project/Phonepe/pulse/data/aggregated/transaction/country/india/state/"

# Listing all the files in the specified directory

Agg_trans_list = os.listdir(path)

print(Agg_trans_list)
```

```python
# Iterating through the list of file names.

for i in Agg_trans_list:
    # Construct the complete file path by combining the directory path and file name
    fpath = path+i+'/'
    print(fpath)
```

Creating DataFrame for Aggregate Transaction 

```python

loc1 ={'State':[], 'Year':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for state in Agg_trans_list:
    state_loc=os.path.join(path, state) + "/"
    agg_trans_year= os.listdir(state_loc)

    for year in agg_trans_year:
        year_loc=state_loc+year+"/"
        agg_trans_files= os.listdir(year_loc)
        
        for file in agg_trans_files:
            file_loc=year_loc+file
            data=open(file_loc,"r")
            A=json.load(data)
            
            for i in A["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                loc1["Transaction_type"].append(name)
                loc1["Transaction_count"].append(count)
                loc1["Transaction_amount"].append(amount)
                loc1["State"].append(state)
                loc1["Year"].append(year)
                loc1["Quarter"].append(int(file.strip(".json")))

agg_trans=pd.DataFrame(loc1)

agg_trans
```

Creating DataFrame for Aggregate User

```python
loc2 = {'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Count': [], 'Percentage': []}

for state in agg_user_list:
    state_loc = path1 + state + "/"
    agg_year_list = os.listdir(state_loc)
    
    for year in agg_year_list:
        year_loc = state_loc + year + "/"
        agg_file_list = os.listdir(year_loc)

        for file in agg_file_list:
            file_loc = year_loc + file
            data = open(file_loc, 'r')
            B = json.load(data)
            try:
                for i in B["data"]["usersByDevice"]:
                    brand_name = i["brand"]
                    counts = i["count"]
                    percents = i["percentage"]
                    loc2["Brands"].append(brand_name)
                    loc2["Count"].append(counts)
                    loc2["Percentage"].append(percents)
                    loc2["State"].append(state)
                    loc2["Year"].append(year)
                    loc2["Quarter"].append(int(file.strip('.json')))
            except:
                pass
agg_user = pd.DataFrame(loc2)

agg_user
```

Creating DataFrame for Map User

```python
loc4 = {"State": [], "Year": [], "Quarter": [], "District": [], "RegisteredUser": [], "AppOpens": []}

for state in map_user_list:
    state_loc = os.path.join(path3, state) + "/"
    map_year_list = os.listdir(state_loc)
    
    for year in map_year_list:
        year_loc = state_loc + year + "/"
        map_file_list = os.listdir(year_loc)
        
        for file in map_file_list:
            file_loc = os.path.join(year_loc, file)
            with open(file_loc, 'r') as data:
                D = json.load(data)
                
                for key, value in D["data"]["hoverData"].items():
                    district = key
                    registereduser = value["registeredUsers"]
                    appOpens = value["appOpens"]
                    loc4["District"].append(district)
                    loc4["RegisteredUser"].append(registereduser)
                    loc4["AppOpens"].append(appOpens)
                    loc4['State'].append(state)
                    loc4['Year'].append(year)
                    loc4['Quarter'].append(int(file.strip('.json')))
                
map_user = pd.DataFrame(loc4)
map_user

```

Creating DataFrame for Top Transaction

```python
loc5 = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Transaction_count': [], 'Transaction_amount': []}

for state in top_trans_list:
    state_loc = path4 + state + "/"
    top_year_list = os.listdir(state_loc)
    
    for year in top_year_list:
        year_loc = state_loc + year + "/"
        top_file_list = os.listdir(year_loc)
        
        for file in top_file_list:
            file_loc = year_loc + file
            data = open(file_loc, 'r')
            E = json.load(data)
            
            for i in E['data']['pincodes']:
                name = i['entityName']
                count = i['metric']['count']
                amount = i['metric']['amount']
                loc5['Pincode'].append(name)
                loc5['Transaction_count'].append(count)
                loc5['Transaction_amount'].append(amount)
                loc5['State'].append(state)
                loc5['Year'].append(year)
                loc5['Quarter'].append(int(file.strip('.json')))
                
top_trans = pd.DataFrame(loc5)
top_trans
```

```python
# Creating agg_trans table
mycursor.execute("create table agg_trans (State varchar(100), Year int, Quarter int, Transaction_type varchar(100), Transaction_count int, Transaction_amount double)")

for i,row in agg_trans.iterrows():
    #here %S means string values 
    sql = "INSERT INTO agg_trans VALUES (%s,%s,%s,%s,%s,%s)"
    mycursor.execute(sql, tuple(row))
    # the connection is not auto committed by default, so we must commit to save our changes
    mydb.commit()
```
